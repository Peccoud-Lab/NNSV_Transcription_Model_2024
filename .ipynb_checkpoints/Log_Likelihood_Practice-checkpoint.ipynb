```python
import numpy as np
from scipy.stats import multinomial

def categorical_log_likelihood(observed_counts, probabilities):
    """
    Calculates the log-likelihood of observed data under a categorical distribution.

    Args:
    observed_counts: A numpy array of observed counts for each category.
    probabilities: A numpy array of probabilities for each category.

    Returns:
    The log-likelihood value.
    """

    return multinomial.logpmf(observed_counts, n=np.sum(observed_counts), p=probabilities)

# Example usage:
observed_counts = np.array([10, 20, 15])
probabilities = np.array([0.3, 0.5, 0.2])

log_likelihood = categorical_log_likelihood(observed_counts, probabilities)
print("Log-likelihood:", log_likelihood)
```
