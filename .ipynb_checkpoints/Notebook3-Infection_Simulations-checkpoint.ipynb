```python
from basico import * #'version 0.36'
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import itertools
import pandas as pd
```

```python
# predict transcription rates for the genes of the 5 gene-shuffled VSV variants used to fit the model

VSV_gene_lengths = np.array([1325, 813, 830, 1664, 6372])

df = pd.read_csv((f"File2-Top_100_Fit_Stats/VSV_fits.csv"))
best_fit_pars = df.iloc[:,3]

data = np.array([4347, 2087, 2348, 1304, 3957]) # VSV gene-shuffled variant replicative fitness data

gene_lengths = np.array([1325, 813, 830, 1664, 6372])
pos_combos = np.array([[1, 4, 2, 3, 5], [1,2,3,4,5], [1,3,2,4,5], [1,3,4,2,5], [1,4,3,2,5]])
gene_lenpos = []
gene_1 = []
for pos_pre in pos_combos:
    pos = pos_pre
    gene_len_temp = [gene_lengths[i-1] for i in pos]
    genome_pos = [np.sum(gene_len_temp[0:i]) + (i-1) * 10 for i in pos]
    gene_1.append(np.min(genome_pos))
    genome_pos = np.array(genome_pos - np.min(genome_pos))
    gene_lenpos.append(genome_pos)

pars = best_fit_pars[0]

my_preds = []
for gene_pos_mn in gene_lenpos:
    pred = [pars**x for x in gene_pos_mn]
    my_preds.append(pred)
         
my_preds = np.array(my_preds)

df = pd.DataFrame(my_preds)


my_preds = np.array(df)
tran_rates = my_preds * base_transcription
```

```python
# load in stochastic model and make a list of the transcription the reactions 

model = load_model("NPMGL_Model_SSA.xml")
set_model_unit(quantity_unit= "#")
set_current_model(model)
base_transcription = get_reactions().loc["transcription_N_0"]["mapping"]["k1"]
n_transcriptions = list(get_reactions().index[np.where(["transcription_N" in x for x in get_reactions().index])[0]])
p_transcriptions = list(get_reactions().index[np.where(["transcription_P" in x for x in get_reactions().index])[0]])
m_transcriptions = list(get_reactions().index[np.where(["transcription_M" in x for x in get_reactions().index])[0]])
g_transcriptions = list(get_reactions().index[np.where(["transcription_G" in x for x in get_reactions().index])[0]])
l_transcriptions = list(get_reactions().index[np.where(["transcription_L" in x for x in get_reactions().index])[0]])

transcription_rxns = [n_transcriptions,p_transcriptions,m_transcriptions,g_transcriptions,l_transcriptions]

```

```python
# Make predictions of replicative fitness for the 5 observed gene-shuffled variants

preds = []
for mod in tran_rates:
    for i in range(len(mod)):
        rxn_rate = mod[i]
        rxn = transcription_rxns[i]
        for j in range(4):
            set_reaction_parameters(f'({rxn[j]}).k1', value=rxn_rate)
    preds_temp = []
    for k in range(1000):
        try:
            tc = basico.run_time_course(duration=57000, start_time=0, stepsize=1000, method="DirectMethod")
            preds_temp.append( tc["V"].loc[57000] )
        except:
            preds_temp.append( np.inf ) 
        if k % 333 == 0:
            print(k, i)
    preds.append(preds_temp)
```

```python
# calculate mean and standard error of predictions
preds_mean = np.mean(preds, axis =1)
preds_se = ( np.std(preds, axis =1) / np.sqrt(1000) ) * 2
```

```python
# graph to compare data
plt.errorbar([0,1,2,3,4], np.array(preds_mean), yerr = preds_se)
plt.plot(np.array(data))
```

```python
# predict transcription rates for the genes of all 120 gene-shuffled VSV variants used to fit the model

pos_combos = list(itertools.permutations([1,2,3,4,5]))
gene_lenpos = []
gene_1 = []
for pos_pre in pos_combos:
    pos = np.array(pos_pre)
    genome_pos = [np.sum(VSV_gene_lengths[pos <= i]) + (i-1) * 10 for i in pos]
    gene_1.append(np.min(genome_pos))
    gene_lenpos.append(np.array(genome_pos))
pars = best_fit_pars[0]

my_preds = []
for gene_pos_mn in gene_lenpos:
    pred = [pars**x for x in gene_pos_mn]
    my_preds.append(pred)
    
df = pd.DataFrame(my_preds)  
    
for i in range(120):  
    df.iloc[i] = (pars**(gene_lenpos[i]- 1325)  )# * df.iloc[i]  
    
df = df/df.iloc[0,0]   
df.columns = ["N", "P", "M", "G", "L"]
df.to_csv("File3-Gene_Shuffled_Variants/transcription_ratios.csv")

my_preds = np.array(df)
tran_rates = my_preds * base_transcription
```

```python
# generate variant names

gene_names = ["N", "P", "M", "G", "L"]
variant_names = ["".join([x for _, x in sorted(zip(pos, gene_names))]) for pos in pos_combos]

```

```python
# Make predictions of replicative fitness for all 120 gene-shuffled variants


preds = []
t = 1
for mod in tran_rates:
    for i in range(len(mod)):
        rxn_rate = mod[i]
        rxn = transcription_rxns[i]
        for j in range(4):
            set_reaction_parameters(f'({rxn[j]}).k1', value=rxn_rate)
    preds_temp = []
    for k in range(5000):
        try:
            tc = basico.run_time_course(duration=57000, start_time=0, stepsize=1000, method="DirectMethod")
            preds_temp.append( tc["V"].loc[57000] )
        except:
            preds_temp.append( np.inf ) 
        if k % 2500 == 0:
            print(k, t)
    t += 1
    preds.append(preds_temp)
```

```python
# calculate mean and standard error of predictions

preds_mean = np.mean(preds, axis =1)
preds_sd = np.std(preds, axis =1)
```

```python
# save predictions

df = pd.concat([ pd.DataFrame(variant_names), pd.DataFrame(tran_rates), pd.DataFrame(preds_mean), pd.DataFrame(preds_sd)], axis=1)
df.columns = ["Variant", "N Transcription Rate (1/s)","P Transcription Rate (1/s)","M Transcription Rate (1/s)",
            "G Transcription Rate (1/s)","L Transcription Rate (1/s)","Mean Virion Production", "SD of Virion Production"]
df.to_csv("File3-Gene_Shuffled_Variants/all_predictions.csv")
```
